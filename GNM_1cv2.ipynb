```python
sqf = calcSqFlucts(gnm[-5:])
```

```python

hinges_cutoff = np.percentile(sqf, 100)

resnames_and_nums = list(zip(calphas1.getResnames(), calphas1.getResnums()))


hinge_residues = [(resname, resnum) for (resname, resnum), value in zip(resnames_and_nums, sqf) if value <= hinges_cutoff]
```

```python
import matplotlib.pyplot as plt
import numpy as np


residue_mapping = {
    "ALA": "A",
    "CYS": "C",
    "ASP": "D",
    "GLU": "E",
    "PHE": "F",
    "GLY": "G",
    "HIS": "H",
    "ILE": "I",
    "LYS": "K",
    "LEU": "L",
    "MET": "M",
    "ASN": "N",
    "PRO": "P",
    "GLN": "Q",
    "ARG": "R",
    "SER": "S",
    "THR": "T",
    "VAL": "V",
    "TRP": "W",
    "TYR": "Y"
}

fig, ax = plt.subplots(figsize=(12, 10), dpi=100)

x_values = np.arange(4, 4 + len(sqf))

ax.plot(x_values, sqf, marker='o', linestyle='-')

subset_color = 'red'

for i, value_tuple in enumerate(hinge_residues):
    if sqf[i] >= 0.01:
        
        value = value_tuple[0]
       
        one_letter_code = residue_mapping.get(value.upper(), "?")
        
        label = f"{one_letter_code} {i+4}"
        ax.annotate(label, (i+4, sqf[i]), textcoords="offset points", xytext=(2,10), ha='center', fontsize=16)
        ax.plot(i+4, sqf[i], marker='o', linestyle='', color=subset_color)

grid(b=True, which='both', axis='both', linewidth=0.08)
x = np.linspace(0, 10, 20)
plt.xticks(np.arange(min(x)+4, max(x)+300, 50))

y = np.linspace(0, 0.02, 10)
plt.yticks(np.arange(min(y), max(y), 0.005))

ax.set_xlabel('Residue Index', fontsize=14)
ax.set_ylabel('Value', fontsize=14)
ax.set_title('Square Fluctuation of last 5 modes of LinB', fontsize=14)

plt.show()
```
