```python
import numpy as np
import matplotlib.pyplot as plt


plt.subplots(figsize=(18, 5), dpi=100)


dfi_values1 = np.load('dfi.npy')
dfi_values2 = np.load('dfi_2qvb.npy')
dfi_values3 = np.load('dfi_7pw1.npy')


absolute_dfi_values1 = np.absolute(dfi_values1)
absolute_dfi_values2 = np.absolute(dfi_values2)
absolute_dfi_values3 = np.absolute(dfi_values3)


x_values1 = np.arange(4, 4 + len(dfi_values1))
x_values2 = np.arange(5, 5 + len(dfi_values2))
x_values3 = np.arange(5, 5 + len(dfi_values3))


plt.plot(x_values1, absolute_dfi_values1, label='DFI LinB')


plt.plot(x_values2, absolute_dfi_values2, label='DFI DmbA')


plt.plot(x_values3, absolute_dfi_values3, label='DFI Ancestor')


plt.legend()


x = np.linspace(0, 10, 20)
plt.xticks(np.arange(min(x)+4, max(x)+300, 7))


y = np.linspace(0, 10, 100)
plt.yticks(np.arange(min(y), max(y), 1))


plt.xlabel('Index')
plt.ylabel('Absolute DFI Value')
plt.title('Dynamic Flexibility Index (DFI) Values')


plt.show()
```
